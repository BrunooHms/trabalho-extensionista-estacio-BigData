**Importando as bibliotecas necessarias**


```python
import numpy as np
import pandas as pd
```

**Carregando o conjunto de dados em um quadro de dados usando Pandas**

```python
df=pd.read_csv('FOOD-DATA-GROUP1.csv')
```

**Explorando o conjunto de dados**

```python
df.info
df.isnull().sum()
```

**Lidar com dados ausentes, incorretos e inválidos**

```python
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan)

df = df.dropna()

df.isnull().sum()

lst = ["Fat", "Saturated Fats", "Monounsaturated Fats", "Polyunsaturated Fats", "Carbohydrates", "Sugars" ]
df[lst] = df[lst].astype("float")
df.dtypes
```

```python
selected_foods = ['caramel custard flan', 'chicken pot pie', 'french fries mcdonalds', 'corn dog', 'cod cooked', 'shrimp fried', 'pepperoni pizza', 'chicken broth', 'chocolate mousse', 'omelet']
```

```python
df_selected = df[df['food'].isin(selected_foods)].copy()
```

```python
df_selected.loc[:, 'food'] = df_selected['food'].replace(translation_dict)
```

```python
df_macronutrients = df_selected[['food', 'Caloric Value', 'Fat', 'Carbohydrates', 'Protein']]
```

```python
df_macronutrients = df_macronutrients.rename(columns={
    'Caloric Value': 'Valor Calórico',
    'Fat': 'Gordura',
    'Carbohydrates': 'Carboidratos',
    'Protein': 'Proteína'
})
```

```python
ax = df_macronutrients.set_index('food').plot(kind='bar', figsize=(14, 8))

plt.title('Valor Calórico e Macronutrientes')
plt.ylabel('Quantidade (g ou kcal)')
plt.xlabel('Alimento')
plt.xticks(rotation=45, ha='right')  # Rotacionando os rótulos do eixo X para melhorar a legibilidade
ax.legend(title='Macronutrientes')

plt.tight_layout()
plt.show()
```

```python
food = df[df['food'] == 'peanut butter']
macronutrients = ['Sodium', 'Polyunsaturated Fats', 'Magnesium', 'Vitamin E', 'Zinc' ]
values = food[macronutrients].values[0]
```

```python
translated_labels = [translation_dict[nutrient] for nutrient in macronutrients]
```

```python
filtered_values = [value for value in values if value > 0]
filtered_labels = [label for value, label in zip(values, translated_labels) if value > 0]
```

```python
plt.figure(figsize=(8, 8))
plt.pie(filtered_values,
        labels=filtered_labels,  # Usando os labels filtrados
        autopct='%1.1f%%',
        colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0'],
        textprops={'fontsize': 10},
        labeldistance=1,
        pctdistance=0.85)

plt.title('Distribuição de Macronutrientes para Manteiga de Amendoim')
plt.show()
```
